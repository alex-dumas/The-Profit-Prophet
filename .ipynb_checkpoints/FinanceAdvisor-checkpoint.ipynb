```python
import csv
import urllib.request
from lxml import html
from yahoo_finance import *
```

```python
with open('us_companies.csv', mode='r') as in_file:
    reader = csv.DictReader(in_file)
    companies = [{
            'company_category': row['company_category'].replace('and', '&')
                                                       .replace(',', '')
            or 'Other',
            'company_name': row['company_name'],
            'description': row['description']
        } for row in reader]
with open('secwiki_tickers.csv', mode='r') as in_file:
    reader = csv.DictReader(in_file)
    name_to_ticker = {row['Name']: row['Ticker'] for row in reader}
```

```python
def get_companies_in_industry(i):
    return list(filter(lambda x: x['company_category'] == i, companies))
```

```python
def get_company_info(c):
    comp = Share(name_to_ticker[c])
    info_comp = filter(lambda x: x['company_name'] == c, companies)
    info = pandas.read_html(urllib.request.urlopen('https://www.google.com/finance?q='+c).read())
    return dict({'history': comp.get_historical(str(datetime.datetime.now().date()\
                                           - datetime.timedelta(days=300)),
                                       str(datetime.datetime.now().date()))}.items() + info_comp)
```
